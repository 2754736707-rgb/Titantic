```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import cross_val_score
from catboost import CatBoostClassifier

# ======================== 1. 加载数据 ========================
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# 分离标签
y = train['Transported'].astype(int)
train.drop('Transported', axis=1, inplace=True)

# 合并以便统一处理
all_data = pd.concat([train, test], axis=0).reset_index(drop=True)
print(f"数据加载完成 train_shape:{train.shape}, test_shape:{test.shape}")

# ======================== 2. 特征工程与缺失值填补 ========================
# 2.1 异常值截断
all_data.loc[all_data.RoomService.gt(9000), 'RoomService'] = 9000
all_data.loc[all_data.FoodCourt.gt(22000), 'FoodCourt'] = 22000
all_data.loc[all_data.ShoppingMall.gt(11000), 'ShoppingMall'] = 11000
all_data.loc[all_data.Spa.gt(17000), 'Spa'] = 17000
all_data.loc[all_data.VRDeck.gt(21000), 'VRDeck'] = 21000

# 2.2 从 PassengerId 提取团队信息
all_data[['TeamId', 'Teamcode']] = all_data['PassengerId'].str.split('_', expand=True).astype(int)

# 2.3 从 Cabin 提取甲板、编号、侧边
all_data[['Deck', 'Num', 'Side']] = all_data['Cabin'].str.split('/', expand=True)

# 2.4 布尔值映射
all_data['CryoSleep'] = all_data['CryoSleep'].map({True: 1, False: 0})
all_data['VIP'] = all_data['VIP'].map({True: 1, False: 0})

# 2.5 家庭/团队大小
all_data['GroupSize'] = all_data['TeamId'].map(all_data['TeamId'].value_counts())
all_data['Name'] = all_data['Name'].fillna('Unknown Unknown')
all_data['Lastname'] = all_data['Name'].str.split().str[-1]
all_data['FamilySize'] = all_data['Lastname'].map(all_data['Lastname'].value_counts())

# 2.6 是否独行
all_data['Is_Solo'] = ((all_data['GroupSize'] == 1) & (all_data['FamilySize'] == 1)).astype(int)

# 2.7 消费相关填补
col_spend = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
all_data.loc[all_data['CryoSleep'] == 1, col_spend] = 0
total_spend_check = all_data[col_spend].sum(axis=1)
all_data.loc[total_spend_check > 0, 'CryoSleep'] = 0

# 总消费额
all_data['Total_Spend'] = all_data[col_spend].sum(axis=1)

# 2.8 VIP 填补
all_data.loc[(all_data.VIP.isna()) & (all_data['Age'] < 18), 'VIP'] = 0
all_data.loc[(all_data.VIP.isna()) & (all_data.HomePlanet.eq('Earth')), 'VIP'] = 0
all_data.loc[(all_data.VIP.isna()) & (all_data.HomePlanet.eq('Mars')) & (all_data.Destination.eq('55 Cancri e')), 'VIP'] = 0
all_data.loc[(all_data.VIP.isna()) & (all_data.Deck.isin(['G', 'T'])), 'VIP'] = 0

# 2.9 HomePlanet 填补
all_data.loc[(all_data.HomePlanet.isna()) & (all_data.VIP.eq(1)) & (all_data.Destination.eq('55 Cancri e')), 'HomePlanet'] = 'Europa'
all_data['HomePlanet'] = all_data.groupby('TeamId')['HomePlanet'].transform(lambda x: x.ffill().bfill())
all_data.loc[(all_data.HomePlanet.isna()) & (all_data.Deck.eq('G')), 'HomePlanet'] = 'Earth'
all_data['HomePlanet'] = all_data.groupby('Lastname')['HomePlanet'].transform(lambda x: x.ffill().bfill())

# 2.10 Age 填补
all_data.loc[(all_data.Age.isna()) & (all_data.VIP.eq(1)), 'Age'] = all_data.loc[(all_data.VIP.eq(1)), 'Age'].median()
all_data.loc[(all_data.Age.isna()) & (total_spend_check > 0), 'Age'] = all_data.loc[(total_spend_check > 0), 'Age'].median()
all_data.loc[(all_data.Age.isna()) & (total_spend_check == 0) & (all_data.CryoSleep.eq(1)), 'Age'] = all_data.loc[(total_spend_check == 0) & (all_data.CryoSleep.eq(1)), 'Age'].median()

# 2.11 Num 转数值并填补
all_data['Num'] = pd.to_numeric(all_data['Num'], errors='coerce')
all_data['Num'] = all_data.groupby('TeamId')['Num'].transform(lambda x: x.fillna(x.median()))

# 2.12 Destination 填补
all_data['Destination'] = all_data['Destination'].fillna(
    all_data.groupby('Lastname')['Destination'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
all_data['Destination'] = all_data['Destination'].fillna(all_data['Destination'].mode()[0])

# 2.13 儿童标志
all_data['Is_Child'] = (all_data['Age'] < 13).astype(int)

# 2.14 基于已有标签的比率特征（需要临时使用 y）
X_temp = all_data.iloc[:len(train)].copy()
full_df = pd.concat([X_temp, y], axis=1).copy()

deck_ratio = full_df.groupby('Deck')['Transported'].mean().to_dict()
all_data['Deck_trans_ratio'] = all_data['Deck'].map(deck_ratio).fillna(full_df['Transported'].mean())

side_ratio = full_df.groupby('Side')['Transported'].mean().to_dict()
all_data['Side_trans_ratio'] = all_data['Side'].map(side_ratio).fillna(full_df['Transported'].mean())

# ======================== 3. 丢弃无用列 + 独热编码 + 迭代填补 ========================
drop_list = ['PassengerId', 'Cabin', 'Name', 'Lastname']
all_data = all_data.drop(drop_list, axis=1)

cols_to_encode = ['HomePlanet', 'Destination', 'Deck', 'Side']
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_df = pd.DataFrame(
    encoder.fit_transform(all_data[cols_to_encode]),
    columns=encoder.get_feature_names_out()
)
all_data = all_data.drop(cols_to_encode, axis=1)
all_data = pd.concat([all_data, encoded_df], axis=1)

# 迭代填补剩余缺失值
it_imp = IterativeImputer(min_value=0)
all_data = pd.DataFrame(it_imp.fit_transform(all_data), columns=all_data.columns)

# ======================== 4. 连续特征分组与标准化 ========================
# 年龄分组
all_data['Age_group'] = pd.cut(
    x=all_data['Age'],
    bins=[-1, 17, 43, all_data['Age'].max()],
    labels=[1, 3, 2]
).astype(float)

# 总消费分组
all_data['Total_Spend_Group'] = pd.cut(
    x=all_data['Total_Spend'],
    bins=[-1, 1, 2250, all_data['Total_Spend'].max()],
    labels=[3, 1, 2]
).astype(float)

# 房间编号分组
all_data['Num_group'] = pd.cut(
    x=all_data['Num'],
    bins=[-1, 300, 690, 1170, all_data['Num'].max()],
    labels=[3, 2, 4, 1]
).astype(float)

# 对部分金额列做 log1p 变换
stand_col = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Total_Spend']
all_data[stand_col] = all_data[stand_col].apply(np.log1p)

# 整体标准化
std_scaler = StandardScaler()
all_scaled = std_scaler.fit_transform(all_data)
all_data = pd.DataFrame(all_scaled, columns=all_data.columns)

# 丢弃不再需要的团队 ID
all_data = all_data.drop(['TeamId', 'Teamcode'], axis=1)

# ======================== 5. 划分训练集与测试集 ========================
X = all_data.iloc[:len(train)].copy()
X_test = all_data.iloc[len(train):].copy()

# ======================== 6. 最终特征列表（与 Notebook 一致） ========================
final_features = [
    'Age', 'VIP', 'RoomService', 'Spa', 'VRDeck', 'Is_Solo', 'Total_Spend',
    'Deck_trans_ratio', 'Side_trans_ratio',
    'HomePlanet_Earth', 'HomePlanet_Mars',
    'Destination_55 Cancri e', 'Destination_TRAPPIST-1e',
    'Deck_C', 'Deck_E', 'Deck_T',
    'Age_group', 'Total_Spend_Group', 'Num_group'
]

# ======================== 7. 模型参数与训练 ========================
params = {
    'depth': 5,
    'iterations': 3000,
    'learning_rate': 0.01,
    'thread_count': -1,
    'verbose': False
}

cat_model = CatBoostClassifier(**params)
cat_model.fit(X[final_features], y)

# 交叉验证评估
scores = cross_val_score(cat_model, X[final_features], y, cv=5)
print("交叉验证准确率: mean = {:.4f}, std = {:.4f}".format(scores.mean(), scores.std()))
print("\n特征重要性:")
print(cat_model.get_feature_importance(prettified=True))

# ======================== 8. 预测并提交 ========================
test_preds = cat_model.predict(X_test[final_features])
output = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Transported': test_preds.astype(bool)
})
output.to_csv('submission.csv', index=False)
print("预测完成，结果已保存为 submission.csv")
```
